# Distribution plots

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 30)
```

```python
sns.jointplot(x='total_bill' ,y='tip' , data=tips, kind='kde') # kind = hec kde reg
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

# Categorical Plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split= True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex') #default jitter = True
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin') #zamena za sve ostale
```

# Matrix plots

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot= True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sns.heatmap(fp, cmap = 'magma', linecolor='white', linewidths=1)
```

```python
sns.clustermap(fp, standard_scale=1)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot,'total_bill')
```

# Regression plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,col='sex', row='time', aspect=0.6, size=8)
```
