```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# functional method
plt.plot(x, y)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
```

```python
plt.subplot(1,2,1) #(broj redova, broj kolona, pozicija)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OOP
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left bottom width height 0-1
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

# Figure size and DPI

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**3, label='x squared')
ax.plot(x,x**2, label='x cubed')
ax.legend(loc=0) # umesto broja od 0-10, moze da se ubaci tuple (x, y)
```

```python
fig.savefig('my_picture.png')
```

#   Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color= '#ff8c00', linewidth=2, linestyle='-') # moze i lw, ls, alpha
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
